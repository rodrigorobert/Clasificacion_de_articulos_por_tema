**Procesamiento**

```python
#vectorizar los articulos

vectorizador = CountVectorizer(max_df=0.8, min_df=2, stop_words='english')
#esto convierte los artículos en sus palabras importantes por frecuencia y por stopword
```

```python
datos = vectorizador.fit_transform(fichero['Articulos'])
```

```python
modelo = LatentDirichletAllocation(n_components=6, random_state=90)

#el 6 son la cantidad de temas en que los quiero agrupar
```

```python
modelo.fit(datos)
```

```python
len(vectorizador.get_feature_names())
#palabras diferntes que ha en contrado
```

```python
temas = modelo.components_
```

```python
len(temas)
```

```python
temas.shape
```

```python
#bucle para recoger las 20 palbras mas probalidad de cada tema

for i,tema in enumerate(temas):
    print(i)
    vector = tema.argsort()[-20:] #(ordena las últimas 20)
    print(vector)
    
    palabras = []
    for palabra in vector:
        texto = vectorizador.get_feature_names()[palabra]
        palabras.append(texto)
    
    print(palabras)
    print('\n')
        
```

```python
#ahora sumaremos una columna con el tema de los artículos... los etiquetaremos

resultados = modelo.transform(datos)

resultados.shape
```

```python
resultados[0].round(2)
#muestra a cual tema tiene más probalidad de pertenecer el artículo que consulto
```

```python
fichero['Tema'] = resultados.argmax(axis=1)
```
